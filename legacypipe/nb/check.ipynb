```python
import os
import numpy as np
from matplotlib import pyplot as plt
from photometry import *
#import settings_north as settings

run = 'north'
output_dir = os.path.join(os.getenv('CSCRATCH'),'legacysim','dr9','legacypipe',run)
randoms_input_fn = '/global/cfs/cdirs/cosmo/data/legacysurvey/dr9/randoms/randoms-1-0.fits'
bricklist_fn = '../bricklist_%s.txt' % run
runlist_fn = '../runlist_%s.txt' % run

def get_bricknames():
    return [brickname[:-len('\n')] for brickname in open(bricklist_fn,'r')]

legacypipe_output_dir = os.path.join('/global/cfs/cdirs/cosmo/data/legacysurvey/dr9/',run)
```

```python
for brick in get_bricknames():
    print('============================= BRICK',brick)
    legacypipe_ref_tractor_fn = os.path.join(legacypipe_output_dir,'tractor',brick[:3],'tractor-%s.fits' % brick)
    #legacypipe_ref_tractor_fn = os.path.join(output_dir,'file0_rs0_skip0','tractor',brick[:3],'tractor-%s.fits' % brick)
    #legacypipe_tractor_fn = os.path.join(output_dir,'tractor',brick[:3],'tractor-%s.fits' % brick)
    #legacysim_tractor_fn = os.path.join(output_dir,'tractor',brick[:3],'tractor-%s.fits' % brick)
    legacysim_tractor_fn = os.path.join(output_dir,'file4_rs0_skip0','tractor',brick[:3],'tractor-%s.fits' % brick)
    legacypipe_ref_tractor = Catalogue.load(legacypipe_ref_tractor_fn)
    legacypipe_tractor = Catalogue.load(legacypipe_tractor_fn)
    legacysim_tractor = Catalogue.load(legacysim_tractor_fn)
    legacypipe_ref_header = Catalogue.load_fits_header(legacypipe_ref_tractor_fn)
    legacypipe_header = Catalogue.load_fits_header(legacypipe_tractor_fn)
    legacysim_header = Catalogue.load_fits_header(legacysim_tractor_fn)
    for header in [legacypipe_ref_header]:
        for key in legacysim_header:
            if key in header and legacysim_header[key] != header[key]:
                print(key,legacysim_header[key],header[key])
    #for tractor in [legacypipe_tractor,legacypipe_ref_tractor]:
    for tractor in [legacypipe_ref_tractor]:
        for field in ['ra','dec','flux_g','flux_r','flux_z']:
            if isinstance(tractor[field][0],np.floating):
                diff = legacysim_tractor[field] - tractor[field]
                if field in ['ra','dec']: diff *= 3600
                div = tractor[field]
                mask = np.abs(div) != 0
                diff = np.abs(diff)[mask]
                print(field, np.max(diff), np.mean(diff))
                #plt.hist(diff[diff<np.percentile(diff,99)])
                #plt.show()
```
